This notebook loads the preprocessed data generated by the mne-bids-pipeline and performs a simple group-level ERP analysis.

```python
import mne
from pathlib import Path

# Define subjects and sessions
subjects = ["sub-02", "sub-06"]
sessions = ["ses-EMS", "ses-Visual", "ses-Vibro"]

# Path to preprocessed data
output_path = Path("./mne_pipeline_output")

for subject in subjects:
    for session in sessions:
        file_path = output_path / subject / f"{session}_cleaned_raw.fif"
        raw = mne.io.read_raw_fif(str(file_path), preload=True)
        print(f"{subject} {session}: {len(raw.ch_names)} channels, Sampling rate: {raw.info['sfreq']} Hz")
```
